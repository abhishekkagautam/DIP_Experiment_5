```python
import cv2
from PIL import Image
from skimage import io
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('test4.png')
```

#Log Transformation

```python
c = 255 / np.log(1 + np.max(img))
log_image = c * (np.log(img + 1))
log_image = np.array(log_image, dtype = np.uint8)
plt.imshow(img)
plt.show()
plt.imshow(log_image)
plt.show()
```

#Power Law

```python
for i in [0.1, 0.5, 1.2, 2.2]:     
    gamma_corrected = np.array(255*(img / 255) ** i, dtype = 'uint8')
    plt.imshow(gamma_corrected)
    plt.show()
```

#Bit Extraction

```python
lst = []
img = cv2.imread('test4.png',cv2.IMREAD_GRAYSCALE)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
         lst.append(np.binary_repr(img[i][j] ,width=8)) 
images=[]
for i in range(8):
  bit_img = (np.array([int(i[0]) for i in lst],dtype = np.uint8) * (2**i)).reshape(img.shape[0],img.shape[1])
  images.append(bit_img)
finalr = cv2.hconcat(images[8:3:-1])
finalv = cv2.hconcat(images[4:0:-1])
 
final = cv2.vconcat([finalr,finalv])
plt.imshow(final)
```

#Intensity Scaling


```python
img = cv2.imread('test4.png',cv2.IMREAD_GRAYSCALE)
row, column = img.shape
img1 = np.zeros((row,column),dtype = 'uint8')
min_range = 10
max_range = 60
for i in range(row):
  for j in range(column):
    if img[i,j]>min_range and img[i,j]<max_range: 
      img1[i,j] = 255
    else: 
      img1[i,j] = 0
plt.imshow(img1)
```
